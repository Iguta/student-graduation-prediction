```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('academic_success.csv')

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Bar plot of graduation rate by international status
# Assuming 'International' is a binary column (1 for international students, 0 for domestic)
# And 'Graduated' is binary as well (1 for graduated, 0 for not graduated)

# Calculate the proportion of students who graduated in each group
graduation_rates = df.groupby('International')['Graduated'].mean()

# Bar plot for graduation rates
plt.figure(figsize=(8, 6))
sns.barplot(x=graduation_rates.index, y=graduation_rates.values)
plt.xticks([0, 1], ['Domestic', 'International'])
plt.ylabel('Graduation Rate')
plt.title('Graduation Rates: Domestic vs International Students')
plt.show()

# Stacked bar plot for proportions of graduated vs not graduated within each group
plt.figure(figsize=(8, 6))
# Create a crosstab for stacked bar plot
stacked_data = pd.crosstab(df['International'], df['Graduated'], normalize='index')

# Plot stacked bar
stacked_data.plot(kind='bar', stacked=True, color=['lightcoral', 'skyblue'], figsize=(8, 6))
plt.xticks([0, 1], ['Domestic', 'International'], rotation=0)
plt.ylabel('Proportion')
plt.title('Graduation vs Non-Graduation: Domestic vs International Students')
plt.legend(['Not Graduated', 'Graduated'])
plt.show()

# If graduation rate is continuous, a box plot would be useful (skip if graduation rate is binary)
# Assuming a continuous 'Graduation Rate' column
if 'Graduation_Rate' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='International', y='Graduation_Rate', data=df)
    plt.xticks([0, 1], ['Domestic', 'International'])
    plt.ylabel('Graduation Rate')
    plt.title('Distribution of Graduation Rates: Domestic vs International Students')
    plt.show()
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Step 2: Load the dataset
# Replace 'students_dropout_success.csv' with the actual path to your dataset
data = pd.read_csv('C:\Intro_DataScience\Week2\predict+students+dropout+and+academic+success\data.csv')

# Step 3: Inspect the dataset
print(data.head())
print(data.info())

# Step 4: Data Preprocessing
# Assuming columns: 'International' (1 for international, 0 for domestic) and 'Graduation_Status' (1 for graduated, 0 for not graduated)
# Remove rows with missing values in 'Graduation_Status'
graduation_data = data.dropna(subset=['Graduation_Status'])

# Step 5: Hypothesis Testing
# Null Hypothesis (H0): International and Domestic students have the same graduation rate.
# Alternative Hypothesis (H1): International students have a different graduation rate compared to domestic students.

# Create a contingency table
contingency_table = pd.crosstab(graduation_data['International'], graduation_data['Graduation_Status'])
print("\nContingency Table:\n", contingency_table)

# Perform Chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"\nChi-squared test results:\nChi-squared value: {chi2}, p-value: {p}\nDegrees of Freedom: {dof}")

# Step 6: Data Visualization - Graduation rate comparison

# Bar Plot: Comparing Graduation Rate between International and Domestic Students
plt.figure(figsize=(10, 6))
sns.countplot(x='International', hue='Graduation_Status', data=graduation_data)
plt.title("Graduation Rate Comparison: International vs Domestic Students")
plt.xlabel("Student Type (0=Domestic, 1=International)")
plt.ylabel("Count of Students")
plt.legend(title="Graduation Status")
plt.show()

# Step 7: Additional Visualization - Percentage graduation rate
graduation_percentage = graduation_data.groupby(['International', 'Graduation_Status']).size().reset_index(name='Count')
graduation_percentage['Percentage'] = graduation_percentage.groupby('International')['Count'].transform(lambda x: x / x.sum() * 100)

plt.figure(figsize=(10, 6))
sns.barplot(x='International', y='Percentage', hue='Graduation_Status', data=graduation_percentage)
plt.title("Graduation Rate Percentage: International vs Domestic Students")
plt.xlabel("Student Type (0=Domestic, 1=International)")
plt.ylabel("Percentage of Students")
plt.legend(title="Graduation Status")
plt.show()
```
