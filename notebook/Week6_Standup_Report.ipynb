```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dictionary for converting numerical values to categories for relevant columns
categorical_mappings = {
    'Marital status': {1: 'Single', 2: 'Married', 3: 'Divorced', 4: 'Widowed'},
    'Daytime/evening attendance': {1: 'Daytime', 0: 'Evening'},
    'Previous qualification': {1: 'Yes', 0: 'No'},
    # Add any other numerical to categorical mappings as needed based on domain knowledge
}

# Convert the numerical columns to categorical labels
for column, mapping in categorical_mappings.items():
    Academic_Success_df[column + '_category'] = Academic_Success_df[column].map(mapping)

# Create the new aggregated column: difference between admission grade and previous qualification grade
Academic_Success_df['Grade Difference'] = Academic_Success_df['Previous qualification (grade)'] - Academic_Success_df.get('Admission grade', 0)

# Set up the plotting environment
plt.figure(figsize=(15, 10))

# 1. Distribution of Grade Difference
plt.subplot(2, 2, 1)
sns.histplot(Academic_Success_df['Grade Difference'], kde=True, color='blue')
plt.title('Distribution of Grade Difference')
plt.xlabel('Grade Difference')

# 2. Boxplot of Grade Difference by Marital Status
plt.subplot(2, 2, 2)
sns.boxplot(x='Marital status_category', y='Grade Difference', data=Academic_Success_df, palette='Set2')
plt.title('Grade Difference by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Grade Difference')

# 3. Scatter plot: Grade Difference vs. Admission grade
plt.subplot(2, 2, 3)
sns.scatterplot(x=Academic_Success_df.get('Admission grade', 0), y='Grade Difference', hue='Marital status_category', data=Academic_Success_df, palette='Set1')
plt.title('Grade Difference vs Admission Grade')
plt.xlabel('Admission Grade')
plt.ylabel('Grade Difference')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'C:/Intro_DataScience/Week4/Academic_Success_Data.csv'
data = pd.read_csv(file_path)

# Check the actual column names to detect any issues
print(data.columns)

# Cleaning column names (removing any extra whitespace or tabs)
data.columns = data.columns.str.strip()

# Dictionary for converting numerical values to categories for relevant columns
categorical_mappings = {
    'Marital status': {1: 'Single', 2: 'Married', 3: 'Divorced', 4: 'Widowed'},
    'Daytime/evening attendance': {1: 'Daytime', 0: 'Evening'},
    'Previous qualification': {1: 'Yes', 0: 'No'},
    # Add any other numerical to categorical mappings as needed based on domain knowledge
}

# Convert the numerical columns to categorical labels
for column, mapping in categorical_mappings.items():
    if column in data.columns:
        data[column + '_category'] = data[column].map(mapping)

# Create the new aggregated column: difference between admission grade and previous qualification grade
data['Grade Difference'] = data['Previous qualification (grade)'] - data.get('Admission grade', 0)

# Set up the plotting environment
plt.figure(figsize=(15, 10))

# 1. Distribution of Grade Difference
plt.subplot(2, 2, 1)
sns.histplot(data['Grade Difference'], kde=True, color='blue')
plt.title('Distribution of Grade Difference')
plt.xlabel('Grade Difference')

# 2. Boxplot of Grade Difference by Marital Status (using hue)
plt.subplot(2, 2, 2)
sns.boxplot(x='Marital status_category', y='Grade Difference', hue='Marital status_category', data=data, palette='Set2', dodge=False)
plt.title('Grade Difference by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Grade Difference')

# 3. Scatter plot: Grade Difference vs. Admission grade
plt.subplot(2, 2, 3)
sns.boxplot(x=data.get('Daytime/evening attendance', 0), y='Grade Difference', hue='Marital status_category', data=data, palette='Set1')
plt.title('Grade Difference vs Admission Grade')
plt.xlabel('Daytime/evening attendance')
plt.ylabel('Grade Difference')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```
