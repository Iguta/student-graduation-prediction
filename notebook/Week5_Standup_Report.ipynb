```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv(r'C:/Intro_DataScience/Week4/Academic_Success_Data.csv')
```

```python
df.columns
```

```python
# Example mapping for categorical variables
marital_status_mapping = {1: 'Single', 2: 'Married', 3: 'Widower', 4: 'Divorced', 5: 'Facto Union', 6: 'Legally Separated'}
df['Marital_Status_Cat'] = df['Marital status'].map(marital_status_mapping)
```

```python
application_mode_mapping = {1: '1st phase - general contingent',
    2: 'Ordinance No. 612/93',
    5: '1st phase - special contingent (Azores Island)',
    7: 'Holders of other higher courses',
    10: 'Ordinance No. 854-B/99',
    15: 'International student (bachelor)',
    16: '1st phase - special contingent (Madeira Island)',
    17: '2nd phase - general contingent',
    18: '3rd phase - general contingent',
    26: 'Ordinance No. 533-A/99, item b2) (Different Plan)',
    27: 'Ordinance No. 533-A/99, item b3 (Other Institution)',
    39: 'Over 23 years old',
    42: 'Transfer',
    43: 'Change of course',
    44: 'Technological specialization diploma holders',
    51: 'Change of institution/course',
    53: 'Short cycle diploma holders',
    57: 'Change of institution/course (International)'}  # Add full mapping
df['Application_Mode_Cat'] = df['Application mode'].map(application_mode_mapping)
```

```python
df
```

```python
### Visualizing the Relationship Between Marital Status and Admission Grade:
plt.figure(figsize=(12, 6))
sns.boxplot(x='Marital_Status_Cat', y='Admission grade', data=df)
plt.title('Admission Grade by Marital Status')
plt.xticks(rotation=45)
plt.show()
```

```python
### Visualizing the Relationship Between Application Mode and Admission Grade:
plt.figure(figsize=(14, 6))
sns.boxplot(x='Application_Mode_Cat', y='Admission grade', data=df)
plt.title('Admission Grade by Application Mode')
plt.xticks(rotation=90)
plt.show()
```
